```python
# 예제 2.3 정부호행렬 여부 판정
import numpy as np
def isPositiveDefinite(mat) :
  row, col = mat.shape
  if (row != col) :
    return False

    for i in range(1, row+1) :
      am = mat[:i, :i]  # 소행렬 작성
      if (np.linalg.det(am) <= 0.0):  # linalg 선형대수학 0보다 작으면 실패
        return False

    return True   # det 0 이상이면 True 반환

if __name__ == '__main__' :
  mat = np.array([[1.0, 2.0, 3.0], [1.0, 2.0, 1.0], [0.0, 1.0, 2.0]])

  if isPositiveDefinite(mat) :  # 정부호행렬 판정
    print('정부호')
  else : 
    print('정부호 아님')
```
